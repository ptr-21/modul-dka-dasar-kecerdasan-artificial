NetworkX adalah library Python yang digunakan untuk membuat, memanipulasi, dan mempelajari struktur dan fungsi dari graf. NetworkX mendukung operasi pada graf tidak berarah, graf berarah, graf berbobot, dan lain-lain.  

# 1. Cara import Library NetworkX

```python
# Import library yang digunakan
import matplotlib.pyplot as plt
import networkx as nx
```

# 2. Membuat Graph

```python
# Fungsi pendukung untuk mencetak graf
def show_graph(G, pos=None, title='') :
  # Membuat pos jika pos tidak ditentukan / tidak diberikan / tidak ditentukan
  if pos == None :
    pos = nx.spring_layout(G)

  # Logika untuk menggambar node
  nx.draw(
      G,                    # Graph
      pos,                  # Posisi node
      with_labels=True,     # Menampilkan nama node
      node_color='red',     # Warna Node
      node_size=2000,       # Ukuran Node
      font_color="white",   # Warna Tulisan/Label pada Node
      font_weight="bold",   # Ketebalan Tulisan/Label pada Node
      width=5               # Ketebalan Garis Edge
  )

  edge_labels = nx.get_edge_attributes(G, 'weight')
  nx.draw_networkx_edge_labels(
      G,
      pos,
      edge_labels=edge_labels,
      font_color='blue',
      font_weight="bold",
      font_size=12,
  )
  plt.margins(0.2)
  plt.title(title)
  plt.show()
```

# 3. Pendefinisian `Graph` kosong

```python
G_undirected = nx.Graph()
# Cetak graf
show_graph(G_undirected, title="Undirected Graph Kosong")
```

```python
G_directed = nx.DiGraph()
# Cetak graf
show_graph(G_directed, title="Directed Graph Kosong")
```

# 5. Menambahkan `Node` pada `Graph`

Terdapat dua cara untuk menambahkan node di NetworkX, yaitu dengan menggunakan `.add_node()` atau dengan menggunakan `.add_nodes_from()`

```python
# Contoh cara add_node()
G_undirected.add_node('A')
G_undirected.add_node('B')
G_undirected.add_node('C')
G_undirected.add_node('D')
G_undirected.add_node('E')
"""#penjelasan G_undirected itu mendefinisikan graph (inisialisasi)
    sedangkan a itu udah ada alamat (address dalam graph)
"""

# Cetak graf
show_graph(G_undirected, pos=pos, title="Add Node Undirected Graph")

# Melihat jumlah node pada graf
print("Jumlah node:", G_undirected.number_of_nodes())
```

```python
# Contoh cara add_nodes_from()
node_data = ['A', 'B', 'C', 'D', 'E']
G_directed.add_nodes_from(node_data)

# Cetak graf
show_graph(G_directed, pos=pos, title="Add Node Directed Graph")

# Melihat jumlah node pada graf
print("Jumlah node:", G_directed.number_of_nodes())
```

# 6. Menambahkan `Edge` pada `Graph`

Untuk menambahkan edge di NetworkX, terdapat 2 cara, yaitu dengan menggunakan `.add_edge()` atau dengan menggunakan `.add_edges_from()`

```python
# Contoh Cara menggunakan add_edge()
G_undirected.add_edge('A', 'B')
G_undirected.add_edge('B', 'C')
G_undirected.add_edge('C', 'D')
G_undirected.add_edge('D', 'B')
G_undirected.add_edge('E', 'D')

# Cetak graf
show_graph(G_undirected, pos=pos, title="Add Edge Undirected Graph")

# Melihat jumlah edge pada graf
print("Jumlah edge:", G_undirected.number_of_edges())
```

```python
# Contoh Cara menggunakan add_edges_from()
edge_data = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'B'), ('E', 'D')]
G_directed.add_edges_from(edge_data)

# Cetak graf
show_graph(G_directed, pos=pos, title="Add Edge Directed Graph")

# Melihat jumlah edge pada graf
print("Jumlah edge:", G_directed.number_of_edges())
```

# 9. Elemen Pada Graph

`.nodes()` Method ini digunakan untuk mengembalikan daftar semua node dalam graf

```python
print("Node pada G_undirected:", G_undirected.nodes())
print("Node pada G_directed:", G_directed.nodes())
```

`.edges()` Method ini digunakan untuk mengembalikan daftar semua edges dalam graf

```python
print("Edge pada G_undirected:", G_undirected.edges())
print("Edge pada G_directed:", G_directed.edges())
```

`.adj()` Method yang digunakan untuk mengakses daftar tetangga atau adjacency dari setiap node. Method Ini memberikan akses langsung ke node-node yang terhubung dengan node tertentu.

```python
print("Tetangga G_undirected")
for node in G_undirected.nodes():
    # Tetangga dari node tertentu
    adj = list(G_undirected.adj[node])
    print(f"Tetangga {node}: {adj}")
```

```python
print("Tetangga G_directed")
for node in G_directed.nodes():
    # Tetangga dari node tertentu
    adj = list(G_directed.adj[node])
    print(f"Tetangga {node}: {adj}")
```

`.degree()` Method ini digunakan untuk mengembalikan derajat dari setiap node, yaitu jumlah total edge yang terhubung ke node tersebut. Pada `undirected graph`, ini adalah jumlah total koneksi, sedangkan pada `directed graph` ini adalah jumlah incoming dan outgoing edges dengan menggunaan method `.in_degree()` dan `.out_degree()`.

```python
print("Derajat node G_undirected")
for node, degree in G_undirected.degree():
    print(f"Derajat {node}: {degree}")

"""Kalo derajat misal saling terhubung garisnya
    misal kalo keluarnya 1 ya 1, kalo 2 ya 2
    tapi kallo terimanya 2 ya 2 (sesuai grafik diatas)
"""
```

```python
print("In-degree node G_directed")
for node, degree in G_directed.in_degree():
    print(f"In-degree {node}: {degree}")
""" In-degree = jumlah node menuju ke node yang dituju
    Out-degree = jumlah node keluar dari node asal
"""
print("Out-degree node G_directed")
for node, degree in G_directed.out_degree():
    print(f"Out-degree {node}: {degree}")
```

# 10. Menambah weight/bobot pada Graph

Weight atau bobot dapat ditambahkan pada edge dalam graf untuk merepresentasikan nilai dari hubungan antara dua node. Terdapat dua metode utama untuk menambahkan weight pada edge. Pertama dengan menggunakan `.add_edge()`, yang digunakan untuk menambahkan satu edge pada graf dan dapat ditambahkan sebagai atribut menggunakan parameter `weight`. Kedua, untuk menambahkan `weight` dapat menggunakan metode `.add_weighted_edges_from()`

```python
# Contoh Cara menggunakan add_edge()
G_undirected.add_edge('A', 'B', weight=10)
G_undirected.add_edge('B', 'C', weight=7)
G_undirected.add_edge('C', 'D', weight=4)
G_undirected.add_edge('D', 'B', weight=8)
G_undirected.add_edge('E', 'D', weight=3)

# Cetak graf
show_graph(G_undirected, pos=pos, title="Add Weight Undirected Graph")
```

```python
# Contoh Cara menggunakan add_weighted_edges_from()
edge_weight_data = [('A', 'B', 10), ('B', 'C', 7), ('C', 'D', 4), ('D', 'B', 8), ('E', 'D', 3)]

G_directed.add_weighted_edges_from(edge_weight_data)

# Cetak graf
show_graph(G_directed, pos=pos, title="Add Weight Directed Graph")
```

# 11. Analisis pada Graph

Dalam NetworkX, berbagai fungsi untuk menganalisis graf disediakan untuk membantu memahami struktur dan karakteristik dari graf yang dibuat.

`.connected_components()` Fungsi ini digunakan untuk menemukan semua connected components dalam undirected graph yang berarti setiap node dalam suatu kelompok dapat dijangkau oleh node lainnya dalam kelompok yang sama.

```python
G = nx.Graph() # Membuat graf baru
# Menambahkan edge (sekaligus node) pada graf
G.add_edges_from([('B', 'D'), ('B', 'E'), ('C', 'A')])

# Membuat kelompok connected components
components = list(nx.connected_components(G))
print(components)

# Cetak graf
show_graph(G, title="Graph Connected Component")
```

`.clustering()`Fungsi ini mengukur clustering coefficient dari setiap node dalam graf. Clustering coefficient adalah ukuran lokal yang menggambarkan seberapa cenderung node-node tetangga dari suatu node saling terhubung satu sama lain.

```python
G = nx.Graph() # Membuat graf baru
# Menambahkan edge (sekaligus node) pada graf
edge_data = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'B'), ('E', 'D')]
G.add_edges_from(edge_data)

# Menghitung clustering coefficient setiap node
clustering_coefficients = nx.clustering(G)
print(clustering_coefficients)

# Cetak graf
show_graph(G, title="Graph Clustering")
```

`.all_shortest_path()` Fungsi ini digunakan untuk menemukan semua jalur terpendek antara dua node dalam graf. Fungsi ini mencari semua rute yang memiliki nilai edge terkecil antara dua node tertentu.

```python
G = nx.Graph() # Membuat graf baru
# Menambahkan edge dan weight (sekaligus node) pada graf
edge_weight_data = [('A', 'B', 10), ('B', 'C', 7), ('C', 'D', 4), ('D', 'B', 8), ('E', 'D', 3)]
G.add_weighted_edges_from(edge_weight_data)

# Membuat daftar jalur terpendek antara semua pasangan node
shortest_paths = dict(nx.all_pairs_shortest_path(G))

# Menampilkan shortest paths dengan format rapi
print("All Shortest Paths:")
for start_node, paths in shortest_paths.items():
    print(f"From {start_node}:")
    for end_node, path in paths.items():
        print(f"  to {end_node}: {' -> '.join(path)}")
    print()

# Menampilkan graf
show_graph(G, title="Graph All Shortest Path")
```

Buatlah graf berbobot `(weighted graf)` dari gambar graf berikut ini menggunakan library NetworkX. Nama simpul harus sesuai dengan nama kota yang tertera pada gambar. Jenis graf yang digunakan adalah undirected graph. Kemudian, dengan menggunakan fungsi shortest path yang ada di NetworkX, carilah semua pasangan shortest path dari graf tersebut!

```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat graf undirected
G = nx.Graph()

# Menambahkan simpul (kota) ke graf
cities = ['None', ]
G.add_nodes_from(cities)

# Menambahkan edge ke graf dengan bobot
edges = [('None', 'None', None), ]
G.add_weighted_edges_from(edges)

# Visualisasi graf
pos = nx.spring_layout(G)  # Posisi untuk Node disetting acak
plt.figure(figsize=(20, 10))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=8, font_weight='bold')
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Weighted Undirected Graph of Romanian Cities")
plt.show()
```

```python
# Mencari shortest path untuk semua pasangan simpul
shortest_paths = dict(nx.all_pairs_shortest_path(None))

# Menampilkan semua pasangan shortest path dan jaraknya
print("\nShortest Paths between all pairs of cities:")
for source in cities:
    for target in cities:
        if source != target:  # Mengabaikan jika source dan target sama
            path = shortest_paths[source][target]
            print(f"Path from {source} to {target}: {path}")
```
