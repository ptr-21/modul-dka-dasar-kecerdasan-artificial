1. Implementasi BFS Pada Tree

```python
import networkx as ns
import matplotlib.pyplot as plt # untuk plot grafik
# Fungsi pendukung untuk mencetak graf
def show_graph(G, pos=None, title='') :
    #meembuat pos jika pos tidak diberikan
    if pos is None:
        pos = ns.spring_layout(G)
    # Fungsi untuk menggambar node
    ns.draw(
        G, # Graf NetworkX
        pos, # Posisi node
        with_labels=True, # Menampilkan nama node
        node_color='red', # Warna node
        node_size=2000, # Ukuran node
        font_color="white", # Warna font label node
        font_weight="bold", # Ketebalan font label node
        width=5 # Ketebalan garis edge 
    )
    edge_labels = ns.get_edge_attributes(G, 'weight')
    # Fungsi untuk menggambar node
    ns.draw_networkx_edge_labels( 
        G,
        pos,
        edge_labels=edge_labels, # Data weight
        font_color='blue', # Warna font label edge
        font_weight="bold", # Berat font label edge
        font_size=12, # Ukuran font label edge
    )
    plt.margins(0.2) # Memberikan margin pada plot
    plt.title(title) # Menampilkan judul graf jika diberikan
    plt.show()
"""membuat Posisi node `pos` """
pos = {
    'A': (0, 3),
    'B': (-10, 2), 'C': (10, 2),
    'D': (-15, 1), 'E': (-5, 1), 'F': (5, 1), 'G': (15, 1),
    'H': (-18, 0), 'I': (-12, 0), 'J': (-8, 0), 'K': (-2, 0),
    'L': (2, 0), 'M': (8, 0), 'N': (12, 0), 'O': (18, 0)
}
G = ns.Graph()
edges = [
    ('A', 'B'), ('A', 'C'),
    ('B', 'D'), ('B', 'E'),
    ('C', 'F'), ('C', 'G'),
    ('D', 'H'), ('D', 'I'),
    ('E', 'J'), ('E', 'K'),
    ('F', 'L'), ('F', 'M'),
    ('G', 'N'), ('G', 'O')
]
"""# Menambahkan edge pada graf"""
G.add_edges_from(edges)
print("Elemen pada Graf:")
print("Daftar Node pada Graf:", G.nodes())
print("Daftar Edge pada Graf:", G.edges())
print("Jumlah Node pada Graf", G.number_of_nodes())
print("Jumlah Edge pada Graf", G.number_of_edges())
show_graph(G, pos=pos)
print("Urutan node yang dikunjungi:")
"""membuat urutan bfs"""
i = 1
for nodes in ns.bfs_edges(G, source='A'):
    print(f"{i}. {nodes[0]} - {nodes[1]}")
i += 1
"""# loop untuk urutan Breadth First Search dengan depth 2"""
print("Urutan node yang dikunjungi:")
i = 1
for nodes in ns.bfs_edges(G, source='A', depth_limit=2):
    print(f"{i}. {nodes[0]} - {nodes[1]}")
    i += 1 
```

2. Implementasi BFS Pada Directed Graph

```python
import networkx as nx  # Perbaikan dari 'network' ke 'networkx'
import matplotlib.pyplot as plt  # Untuk plot grafik

# Fungsi untuk menampilkan graf
def show_graph(G, pos=None, title='', labels=None):
    if pos is None:
        pos = nx.spring_layout(G)

    # Menggambar graf
    nx.draw(
        G,
        pos,
        with_labels=True,
        labels=labels,
        node_color='red',
        node_size=2000,
        font_color="white",
        font_weight="bold",
        width=5
    )

    edge_labels = nx.get_edge_attributes(G, 'weight')
    
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_color='blue',
        font_weight="bold",
        font_size=12
    )

    plt.margins(0.2)
    plt.title(title)
    plt.show()

# Membuat posisi node manual
pos = {
    'A': (0, 1),
    'B': (1, 2),
    'C': (1, 0),
    'D': (2, 3),
    'E': (2, 1.5),
    'F': (3, 1),
}

# Membuat graf terarah
G = nx.DiGraph()
edges = [
    ('A', 'B'), ('A', 'C'),
    ('B', 'D'), ('B', 'E'),
    ('C', 'E'), ('C', 'F'),
    ('E', 'F'), ('E', 'D')
]


G.add_edges_from(edges)

# Label node, termasuk label khusus Start dan Goal
labels = {node: node for node in G.nodes()}
labels['A'] = 'A\nStart'
labels['F'] = 'F\nGoal'

# Menampilkan informasi graf
print("Elemen pada Graf:")
print("Daftar Node pada Graf:", list(G.nodes()))
print("Daftar Edge pada Graf:", list(G.edges()))
print("Jumlah Node pada Graf:", G.number_of_nodes())
print("Jumlah Edge pada Graf:", G.number_of_edges())

# Tampilkan graf
show_graph(G, pos=pos, title='Contoh Graf Terarah', labels=labels)

# BFS tanpa batas kedalaman
print("\nUrutan node yang dikunjungi (tanpa depth_limit):")
i = 1
for nodes in nx.bfs_edges(G, source='A'):
    print(f"{i}. {nodes[0]} - {nodes[1]}")
    i += 1

# BFS dengan batas kedalaman 1
print("\nUrutan node yang dikunjungi (depth_limit = 1):")
i = 1
for nodes in nx.bfs_edges(G, source='A', depth_limit=1):
    print(f"{i}. {nodes[0]} - {nodes[1]}")
    i += 1
```
