memberi label dalam 1 figure dan banyak plot

Memberikan grafik dalam plot atau figure

```python
# Membuat figure dan axis
fig, ax = plt.subplots(figsize=(8,6))  

# Membuat plot sederhana
ax.plot([1,2,3,4], [10,20,25,30], label='Data1')

# Menambah label dan judul
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Contoh Plot dalam Figure")

# Menampilkan legenda
ax.legend()

# Menampilkan figure
plt.show()
```

Legend Plot

```python
fig = plt.figure()
plt.plot([1, 2, 3, 4, 5, 6], label="Data 1")
plt.plot([2, 3, 5, 6, 7, 8], label="Data 2")
# Menampilkan legenda
plt.legend()
# Menampilkan figure
plt.show()
```

Memberikan Anotasi Pada Label

```python
fig = plt.figure()
# Membuat linechart pada
plt.plot([1, 2, 3, 4, 5, 6], label="Data 1")
plt.plot([2, 3, 5, 6, 7, 8], label="Data 2")
# Membuat anotasi pada data
plt.annotate("Poin Penting", xy=(2, 5), xytext=(3, 7),
    arrowprops=dict(facecolor='black', shrink=0.05))
# Menampilkan legenda berdasarkan label pada data
plt.legend()
# Menampilkan figure
plt.show()
```

Penggunaan Labeling dalam Plot

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [12, 18 , 30, 40]
# Membuat plot
fig = plt.figure(figsize=(10,6))
plt.plot(x, y1, label="Contoh Data 1")
plt.plot(x, y2, label="Contoh Data 2")
# Menambahkan judul dan label
plt.title("Contoh Plot dengan Labeling")
plt.xlabel("Data X")
plt.ylabel("Data Y")
# Menambahkan legenda
plt.legend()
# Menampilkan figure
plt.show()
```

Plot Pairwise Pada Matplotlib

1. Plot Garis (`plt.plot()`)

```python
# Data
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [12, 18, 30, 40]
# Membuat figure
fig = plt.figure(figsize=(8,6))
# Membuat linechart dengan plt.plot()
plt.plot(x, y1, color="green", linewidth=3, label="Data Y1")
plt.plot(x, y2, linestyle='--', label="Data Y2")
plt.title("Contoh Plot Garis")
plt.xlabel("Data X")
plt.ylabel("Data Y")
# Menampilkan figure
plt.legend()
plt.show()
```

2. Scatter Plot (`plt.scatter()`)

```python
# Data
x = [1, 2, 3, 4, 5, 6, 7 ,8]
y1 = [10, 21, 25, 30, 34, 41, 47, 50]
y2 = [12, 20, 26, 29, 33, 37, 41, 43]
# Membuat figure
fig = plt.figure(figsize=(10,5))
# Membuat scatter plot dengan plt.scatter()
plt.scatter(x, y1, s=100, c='red', alpha=0.65, label='Data 1')
plt.scatter(x, y2, s=50, c='blue', alpha=0.65, label='Data 2')
# Labeling plot
plt.title("Contoh Scatter Plot")
plt.xlabel("Data X")
plt.ylabel("Data Y")
# Menampilkan figure
plt.legend()
plt.show() 
```

Bar Plot (`plt.bar()`)

```python
kategori = ['A', 'B', 'C', 'D'] # Label setiap kategori
nilai = [10, 20, 25, 30] # Nilai setiap kategori
# Membuat Figure
fig = plt.figure(figsize=(6,8))
# Membuat bar plot dengan plt.bar()
plt.bar(kategori, nilai, hatch='/', color='skyblue')
# Labeling plot
plt.title("Contoh Bar Plot")
plt.xlabel("Kategori")
plt.ylabel("Nilai")
# Menampilkan figure
plt.show()
```

Fill Between (`plt.feel_between()`)

```python
# Data
x = [1, 2, 3, 4]
y1 = [9, 20, 27, 35]
y_main = [7, 18, 22, 29] # Data garis utama
y2 = [6, 15, 18, 23]
# Membuat figure
fig = plt.plot(figsize=(10,6))
# Membuat fill_between plot
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.5, label='Fill Plot')
# Membuat plot garis utama
plt.plot(x, y_main, color='blue', linewidth=2, label='Garis Utama')
# Labeling plot
plt.title("Fill Between")
plt.xlabel("X")
plt.ylabel("Y")
# Menampilkan figure
plt.legend(loc='lower right')
plt.show()
```

Stack Plot (`plt.stackplot()`)

```python
# Data
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 10, 15, 20]
y3 = [2, 5, 8, 12]
# Membuat figure
fig = plt.figure(figsize=(10,6))
# Membuat stack plot dengan .stackplot()
plt.stackplot(x, y1, y2, y3, labels=['Y1', 'Y2', 'Y3'])
# Labeling plot
plt.title("Stack Plot")
plt.xlabel("X")
plt.ylabel("Y")
# Menampilkan figure
plt.legend(loc='upper left')
plt.show()
```

Plot Distribusi Statistik Pada Matplotlib

1. Histogram (`plt.hist()`)

```python
data = [1,9,2,2,8,8,7,7,7,3,3,3,4,4,4,4,6,6,6,6,5,5,5,5,5,5]
# Membuat plot
fig = plt.figure(figsize=(10,6))
# Membuat histogram dengan .hist()
plt.hist(data, bins=9, color="blue", alpha=0.7)
# Labeling plot
plt.title("Contoh Histogram")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
# Menampilkan figure
plt.show()
```

2. Box Plot (`plt.boxplot()`)

```python
data = [7, 5, 8, 9, 6, 7, 5, 8, 6, 7, 6, 8, 9, 4, 7, 5, 9, 8,
6, 7]
# Membuat plot
fig = plt.figure(figsize=(8,6))
# Membuat box plot dengan .boxplot()
plt.boxplot(data, patch_artist=True, widths=0.50)
# Labeling Plot
plt.title("Contoh Box Plot")
plt.ylabel("Nilai")
# Menampilkan figure
plt.show()
```

3. Violin Plot (`plt.violinplot()`)

```python
# Data
data = [7, 5, 8, 9, 6, 7, 5, 8, 6, 7, 6, 8, 9, 4, 7, 5, 9, 8, 6, 7]
# Membuat Plot
fig = plt.figure(figsize=(6,8))
# Membuat violin plot dengan .violinplot()
plt.violinplot(data)
# Labeling Plot
plt.title("Contoh Violin Plot")
plt.ylabel("Nilai")
# Menampilkan figure
plt.show()
```

4. Pie Chart (`plt.pie()`)

```python
label = ['A', 'B', 'C', 'D', 'E']
nilai = [15, 25, 35, 20, 5]
# Membuat Plot
fig = plt.figure(figsize=(8,8))
# Membuat pie chart dengan .pie()
plt.pie(nilai, labels=label, autopct='%1.1f%%')
# Labeling plot
plt.title("Contoh Pie Chart")
# Menampilkan figure
plt.legend()
plt.show()
```
